```python
import datetime as dt

nascimento_ano = int(input("Em qual ano voce nasceu? "))
nascimento_mes = int(input("Em qual mês voce nasceu (1-12)? "))
nascimento_dia = int(input("Em qual dia voce nasceu (1-31)? "))

hoje = dt.date.today()

idade = hoje.year - nascimento_ano

if (hoje.month, hoje.day) < (nascimento_mes, nascimento_dia):
    idade -= 1

print(f"Sua idade atual é {idade} anos")
```

```python
import datetime as dt

while True:
    try:
        data_str = input("Digite sua data de nascimento (DD/MM/YYYY): ")
        nascimento = dt.datetime.strptime(data_str, "%d/%m/%Y").date()
        break
    except ValueError:
        print("Data inválida! Use o formato DD/MM/YYYY.")

hoje = dt.date.today()

idade = hoje.year - nascimento.year

if (hoje.month, hoje.day) < (nascimento.month, nascimento.day):
    idade -= 1

print(f"Sua idade atual é {idade} anos")
```

```python
#Exercício 05:Faça um programa para a leitura de duas notas parciais de um aluno. O programa deve calcular a média alcançada por aluno e apresentar:
  #A mensagem "Aprovado", se a média alcançada for maior ou igual a sete;
  #A mensagem "Reprovado", se a média for menor do que sete;
  #A mensagem "Aprovado com Distinção", se a média for igual a dez.

n1 = float(input("Qual sua primeira nota :"))
n2 = float(input("Qual sua segunda nota :"))
media = (n1+n2)/2

if media >= 7:
  print("Aprovado")
elif media >= 9.9:
  print("Aprovador com Distinção")
else:
  print("Reprovado")
print(f"Sua nota media foi {media:.2f}")

```

```python
#Nivel Hard - cada tentativa um novo numero:
import random
n = random.randrange(101)
while tentativa != n:
  tentativa = int(input("Informe um numero de 1 a 100: "))
  if tentativa < n:
    print("Muito baixo!")
  elif tentativa > n:
    print("Muito alto!")
  else:
    print("Acertou!")
    break
```

```python
#Faça um programa que conte a quantidade de consoantes e a quantidade de vogais de uma palavra
  #Ex: Brasil
  #Quantidade  de vogal = 2
  #Quantidade  de consoantes = 4

palavra = input("Me informe uma palavra: ").lower().strip()

vogal = ["a", "e", "i", "o", "u"]
n_vogais = 0
n_consoantes = 0

for letra in palavra:
    if letra in vogal:
        n_vogais += 1
    elif letra.isalpha():
        n_consoantes += 1

print("Quantidade de vogais =", n_vogais)
print("Quantidade de consoantes =", n_consoantes)
```
